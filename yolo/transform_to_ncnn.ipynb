```python
from ultralytics import YOLO
import os
import shutil

IMGSZ   = 416
NAME    = "combined_apple_2_null"
NCNN_SAVE_DIR = "ncnn_models"  # NCNN 모델 저장 폴더


best_pt_path = r"runs\detect\combined_apple_2_null\yolo11n\weights\best.pt"
print(f"[INFO] best.pt 경로: {best_pt_path}")

# ─────────────────────────────────────────────
# 3. NCNN 변환
# ─────────────────────────────────────────────
best_model = YOLO(best_pt_path)
best_model.export(
    format="ncnn",
    imgsz=IMGSZ,
    half=False   # 라즈베리파이는 FP16 미지원이라 False
)
```

```python
# ─────────────────────────────────────────────
generated_ncnn = best_pt_path.replace(".pt", "_ncnn_model")  # ultralytics 자동 생성 경로
os.makedirs(NCNN_SAVE_DIR, exist_ok=True)
target_ncnn = os.path.join(NCNN_SAVE_DIR, f"{NAME}_ncnn_model")

if os.path.exists(target_ncnn):
    shutil.rmtree(target_ncnn)  # 기존 폴더 덮어쓰기

shutil.move(generated_ncnn, target_ncnn)
print(f"[INFO] NCNN 모델 저장 완료: {target_ncnn}")
```
