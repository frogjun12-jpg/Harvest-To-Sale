```python
import sys, torch
from ultralytics import YOLO
from roboflow import Roboflow
print(sys.executable)
print(torch.__version__)
print(torch.version.cuda)
print(torch.cuda.is_available())
```

```python
# 1. 모델 로드 (YOLO11m 모델 사용)
model = YOLO("yolo26n.pt")

# 2. 모델 학습
results = model.train(
    data="combined_apple_2_null/data.yaml", # 다운로드된 data.yaml 경로
    epochs=120,                                # 학습 반복 횟수
    imgsz=416,                                 # 이미지 사이즈
    batch=16,                                  # 배치 사이즈 (메모리 사양에 따라 조절)
    project="testtest",               # 결과 저장 프로젝트명
    name="yolo26n",       # 실험 이름
    device='0',
    patience=25# GPU 사용 시 0, CPU 사용 시 'cpu'
)
```

```python
### combined_appled version 2
metrics = model.val(
    data="./test_for_combined_apple_2_damage/data.yaml",
    split="val"
)
```
